# Introduction

This dataset provide a comprehensive analysis of the online sales data for a diverse range of products,their various category,regions and payment method.The dataset include transactionrecorded from 2024,capturing crucial detils such as product categories,units sold,pricing and regional sales distribution

# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Load Data

```python
data=pd.read_csv('Online Sales Data.csv')
data
```

# About Data

> Column Description
  * Transaction ID:Unique id for each transaction
  * Date:The date on which the transaction happened
  * Product Category:The categoey of the purchased project.(Electronics,Home Appliances,Clothing,Books,Beauty Products,Sports)
  * Product Name:The names of specific product.
  * Units Sold:The quantity of units sold in ech transaction
  * Unit Price:The price of per unit of product
  * Total Revenue:The revenue generated from each transaction
  * Region:The region were sales happen.(North America,Europe,Asia)
  * Payment Method:The method used by customers to complete transaction.(Credit Card,PayPal,Debit Card))

# Extracting Year,Month and Day

```python
data['Year']=data['Date'].dt.year
data['Month']=data['Date'].dt.month_name()
data['Day']=data['Date'].dt.day_name()
```

```python
data
```

```python
# 1.Total revenue of product categories.
data.groupby(['Product Category'])['Total Revenue'].sum().plot(kind='bar',color='brown')
plt.title('Total revenue of product category')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

* This plot show dominanace of the Electronics category in revenue generation,towering above all other categories with an impressive total revenue 35000 which indicate strong demand for electronic products in online sales market
* books and beauty product generate least revenue indicating the weak demand of those category which may be due to low price.

```python
# 2.Top 10 Best Selling Product:
data.groupby(['Product Name'])['Total Revenue'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top Ten Best Selling Products')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.xticks(rotation=90)
plt.show()
```

* The bar chart highlights the top 10 best-selling products by revenue.
* The Canon EOS R5 Camera stands out as leading product,generating highest revenue indicating its strong market demand.
* Samsung Odyyssey G9 Gaming Monitor ranks 10th  genearated least revenue.

```python
# 3.Region generating highest revenue
data.groupby(['Region'])['Total Revenue'].mean().plot(kind='bar')
plt.title('Avearge Regional Revenue')
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.xticks(rotation=0)
plt.show()
```

* This plot shows average revenue generated by each region.
* North America leads with highest average revenue suprassing both Asia and Europe.
* This shows the demand of online sales market in north america.

```python
# 4.revenue trend by month
data.groupby(['Month'])['Total Revenue'].sum().plot(kind='line',marker='o',color='red')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Revenue Trend')
plt.xticks(rotation=45)
plt.show()
```

* Maximum revenue has generated in january as compared to other month
* least revenue was genearated in July and August.

```python
# 5.count of unit sold:
data.groupby(['Product Category'])['Units Sold'].sum().plot(kind='bar',color='')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.title(' Total Unit Sold by Product Category')
plt.show()
```

* maximum units were sold under clothing category .
  

```python
#6.variation of revenue in different region by product category:
sns.barplot(x='Region',y='Total Revenue',data=data,hue='Product Category')
plt.title('Revenue Generated by Region')
plt.show()
```

* In North America , the revenue generating from Electronics were dominating,but there no revenue from Home Appliances,Clothing,Beauty Products and sports
* similarly, in Europe and Asia ,some category was generating revenue others doesn't


```python
# 7.percentage of product category to total revenue:
data.groupby(['Product Category'])['Total Revenue'].sum().plot(kind='pie',autopct='%1.1f%%',explode=(0,0,0,0.1,0,0),shadow=True)
plt.title('Distribution of Product Categoty')
plt.show()
```

* Electronics is dominating with 43.4% of Total Revenue where as books contribute least revenue of 2.3% to the total revenue.

```python
# 8.Number of units sold over time:
data.groupby(['Month'])['Units Sold'].sum().plot(kind='line',marker='*',color='g')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Total Units Sold Over Month')
plt.show()
```

* maximum number of units were sold in march and least number of units were sold in July and August.

```python
#9.distribution of region by sales:
data.groupby(['Region'])['Total Revenue'].sum().plot(kind='pie',autopct='%1.2f%%')
plt.title('Distribution of Revenue over Region')
plt.show()
```

* Maximum revenue was generated in North America

```python
# 10.revenue trend of top selling product category:
a=data[data['Product Category']=='Electronics']
sns.lineplot(x='Month',y='Total Revenue',data=a)
plt.title('Total revenue of Electronics over Month')
plt.show()
```

* maximum revenue is generated in april 

```python
# 11. count  of payement method:
data['Payment Method'].value_counts().plot(kind='pie',autopct='%1.2f%%',explode=(0.1,0,0),startangle=45,shadow=True,colors=['red','yellow','skyblue'])
plt.title('Distribution of payment method')
plt.show()
```

* most the payment was done by credit card

```python
# 12.Distribution of product category :
sns.lineplot(x='Product Category',y='Unit Price',data=data,hue='Payment Method')
plt.xticks(rotation=45)
plt.title('distibution of product category with payment method')
plt.show()
```

* mainly used payment method is Credit Card for Electronics category
* Debit Card is rarely used and is for clothing category

```python
#13.average unit price by category:
data.groupby(['Product Category'])['Unit Price'].mean().plot(kind='bar',color='indigo')
plt.title('Average Unit Price of Product Category')
plt.xlabel('Category')
plt.ylabel('Price')
plt.show()
```

* electronics has generated maimum average revenue

```python
# 14.sales differ by days of week:
data.groupby(['Day'])['Units Sold'].sum().plot(kind='line',marker='o',color='violet')
plt.title('Daily Sales Trend')
plt.show()
```

* most of the sales was done in saturday and least number of sales happen in wednesday

```python
#15.Revenue Generated by top 10 products:
data.groupby(['Product Name'])['Total Revenue'].sum().sort_values(ascending=False).head(10).plot(kind='pie',autopct='%1.2f%%')
plt.title('Revenue Generated by top 10 products')
plt.show()
```

* Canon EOS R5 camera has generated more revenue.

```python
#16.Price Distribution over Region
sns.lineplot(x='Product Category',y='Unit Price',data=data,hue='Region')
plt.xticks(rotation=45)
plt.title('Price Distribution over Region')
plt.show()
```

* This shows that North America has highest price for electronics but othere is a price drop in other category.
  

```python
# 17.correlation between total revenue ,region:
corr=data[['Total Revenue','Unit Price']].corr()
sns.heatmap(corr,annot=True)
plt.title('correlation between Total Revenue and Unit Price')
plt.show()
```

* here total revenue and unit price has astrong positive relationship.
* when unit price increases total revenue also increase

```python
# 18.average units sold in each day of a week:
data.groupby(['Day'])['Units Sold'].mean().plot(kind='bar',color='r')
plt.title('Avearge Unit Sold')
plt.xlabel('Day')
plt.ylabel('Unit Sold')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

* most of the products weresold in saturday
* comparatively least number of products were sold in wednesday

```python
# 19. to check outliers in total revenue:
sns.boxplot(y='Total Revenue',data=data,x='Day')
plt.title('Outliers of Total Revenue over Day')
plt.xticks(rotation=45)
plt.show()
```

* there is outliers present in each day
* the median is almost same in all days


```python
# 20.total units sold over day:
data.groupby(['Day'])['Units Sold'].sum().plot(kind='bar',color='pink')
plt.title('Total Units Sold over Day')
plt.xlabel('Day')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

* more units were sold in saturday
